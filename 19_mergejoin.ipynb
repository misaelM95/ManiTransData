```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({'A':['A0','A1','A2','A3'],
 'B':['B0','B1','B2','B3'],
 'C':['C0','C1','C2','C3'],
 'D':['D0','D1','D2','D3'],})

df2 = pd.DataFrame({'A':['A4','A5','A6','A7'],
 'B':['B4','B5','B6','B7'],
 'C':['C4','C5','C6','C7'],
 'D':['D4','D5','D6','D7'],})
```

```python
#Aquí se aprecia el índice tomado por cada DF
pd.concat([df1,df2])
```

```python
#Aquí se ignora el index
pd.concat([df1,df2],ignore_index=True)
```

```python
#No es muy común realizar una concatenación así, lo común es hacer las concatenaciones por axis=0
pd.concat([df1,df2],axis=1)
```

MERGE

```python
izq = pd.DataFrame({'key':['k0','k1','k2','k3'],
 'A':['A0','A1','A2','A3'],
 'B':['B0','B1','B2','B3']})

der = pd.DataFrame({'key':['k0','k1','k2','k3'],
 'C':['C0','C1','C2','C3'],
 'D':['D0','D1','D2','D3']})
```

```python
#Hace una relación con las columnas que tienen en común, en este caso la Col 'Key'
#Aparte se lleva las columnas que no son iguales
izq.merge(der, on='key')
```

```python
izq = pd.DataFrame({'key':['k0','k1','k2','k3'],
 'A':['A0','A1','A2','A3'],
 'B':['B0','B1','B2','B3']})

der = pd.DataFrame({'key_2':['k0','k1','k2','k3'],
 'C':['C0','C1','C2','C3'],
 'D':['D0','D1','D2','D3']})
```

```python
#Aquí generá error porqué no se encuentra en común una columna
izq.merge(der, on='key')
```

```python
#Se puede aplicar estos comandos para que se pueda hacer la relación solicitada
izq.merge(der, left_on='key', right_on='key_2')
```

```python
izq = pd.DataFrame({'key':['k0','k1','k2','k3'],
 'A':['A0','A1','A2','A3'],
 'B':['B0','B1','B2','B3']})

der = pd.DataFrame({'key_2':['k0','k1','k2',np.nan],
 'C':['C0','C1','C2','C3'],
 'D':['D0','D1','D2','D3']})
```

```python
der
```

```python
#Aquí solo aplica un Inner Join, es decir, solo nos arroja los datos que coinciden
izq.merge(der, left_on='key', right_on='key_2', how='left')
```

```python
izq.merge(der, left_on='key', right_on='key_2', how='right')
```

# JOIN - INDEX MATCH

```python
izq = pd.DataFrame({'A':['A0','A1','A2','A3'],
 'B':['B0','B1','B2','B3']},
index=['K0','K1','K2','K3'])

der = pd.DataFrame({'C':['C0','C1','C2','C3'],
 'D':['D0','D1','D2','D3']},
index=['K0','K2','K4','K5'])
```

```python
#Aquí afecta directamente a los índex
izq.join(der)
```

```python
#Aquí solo nos arroja los valores que hacen match con los Index (k's)
izq.join(der,how='inner')
```

```python
#Aquí nos arroja toda la información que tenemos en los DF
izq.join(der,how='outer')
```
