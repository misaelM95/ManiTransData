```python
import pandas as pd
```

```python
df_books = pd.read_csv('/content/drive/MyDrive/MAN&TRANSDATA/bestsellers-with-categories.csv')
df_books.head(2)
```

```python
df_books.groupby('Author').count()
```

```python
df_books.groupby('Author').mean()
```

```python
df_books.groupby('Author').min()
```

```python
df_books.groupby('Author').max()
```

```python
#Al aplicar el groupby se convierte la columna, en este caso "Author", nuestro índice del DF.
df_books.groupby('Author').sum()
```

```python
#LOC = Un buscador de índices por el label
df_books.groupby('Author').sum().loc['William Davis']
```

```python
#Con reset_index convierte el índice, la columna "Author", en columna
df_books.groupby('Author').sum().reset_index()
```

```python
#El índex es la columna 'Author' y las demás columnas nos arrojan el min y max
df_books.groupby('Author').agg(['min','max'])
```

```python
df_books.groupby('Author').agg({'Reviews':['min','max'], 'User Rating':'sum'})
```

```python
df_books.groupby(['Author','Year']).count()
```
